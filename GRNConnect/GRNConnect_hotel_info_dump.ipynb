```python
import requests

url = "https://api-sandbox.grnconnect.com/api/v3/hotels/?country=IN"

payload = {}
headers = {
    'API-key': 'cda7a3d1a85a048030eca511a2805c59',
    'Accept': 'application/json',
    'Content-Type': 'application/json'
}

response = requests.request("GET", url, headers=headers, data=payload)

if response.status_code == 200:
    response_data = response.json()

    # Extract hotel data
    hotels = response_data.get("hotels", [])

    # Specify SQL file name
    sql_file = "hotels_data.sql"

    # Open the SQL file for writing
    with open(sql_file, mode='w', encoding='utf-8') as file:
        for hotel in hotels:
            # Remove 'description' if it exists in hotel data
            if 'description' in hotel:
                del hotel['description']

            # Prepare SQL INSERT INTO statement for each hotel
            columns = ', '.join(hotel.keys())  # Column names (keys)
            values = ', '.join([f"'{str(value).replace('\'', '\\\'')}'" for value in hotel.values()])  # Values with proper escaping

            # Generate SQL statement
            sql_statement = f"INSERT INTO hotels ({columns}) VALUES ({values});\n"
            file.write(sql_statement)

    print(f"Data has been saved to {sql_file}")
else:
    print(f"Error: {response.status_code}")
    print(response.text)
```

```python
import requests
import csv

def read_country_codes(file_path):
    with open(file_path, 'r') as file:
        return [line.strip() for line in file.readlines()]

# Function to fetch hotel data for a given country
def fetch_hotels_data(country_code, headers):
    url = f"https://api-sandbox.grnconnect.com/api/v3/hotels/?country={country_code}"
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        return response.json().get("hotels", [])
    else:
        print(f"Error fetching data for country {country_code}: {response.status_code}")
        print(response.text)
        return []

# Function to write hotel data into a CSV file
def write_hotels_to_csv(hotels, csv_file):
    with open(csv_file, mode='a', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)

        for hotel in hotels:
            if 'description' in hotel:
                del hotel['description']

            if not file.tell():  # Check if the file is empty
                writer.writerow(hotel.keys())  # Write header row

            writer.writerow(hotel.values())

# Function to fetch data for all countries and save to CSV file
def fetch_and_save_hotels_data(country_codes_file, csv_file, headers):
    country_codes = read_country_codes(country_codes_file)
    for country_code in country_codes:
        hotels = fetch_hotels_data(country_code, headers)
        if hotels:
            write_hotels_to_csv(hotels, csv_file)
            print(f"Data for country code {country_code} has been saved to {csv_file}")
        else:
            print(f"No data for country code {country_code}.")

    print(f"All data has been saved to {csv_file}")

# Main block of code
if __name__ == "__main__":
    # Headers for API request
    headers = {
        'API-key': 'cda7a3d1a85a048030eca511a2805c59',
        'Accept': 'application/json',
        'Content-Type': 'application/json'
    }

    # Path to the file containing country codes
    country_codes_file = 'D:/hotels_content_to_create_json_file/GRNConnect/country_codes.txt'

    # CSV file to save hotel data
    csv_file = "hotels_data.csv"

    # Fetch and save hotel data
    fetch_and_save_hotels_data(country_codes_file, csv_file, headers)
```
