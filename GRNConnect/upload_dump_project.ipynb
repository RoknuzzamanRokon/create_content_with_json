```python
import pandas as pd
from sqlalchemy import create_engine, MetaData, Table
import numpy as np

DATABASE_URL = "mysql+pymysql://root:@localhost/csvdata01_02102024"
engine = create_engine(DATABASE_URL)

chunk_size = 1000
file = "D:/Rokon/hotels_content_to_create_json_file/GRNConnect/hotels_data.csv"

metadata = MetaData()
grnc_table = Table('grn_connect', metadata, autoload_with=engine)

def handle_nan_values(data):
    """Convert NaN and missing values to None."""
    for key, value in data.items():
        if isinstance(value, float) and (pd.isna(value) or np.isnan(value)):
            data[key] = None
    return data

def preprocess_csv(file_path, expected_columns):
    """Fix rows with inconsistent column counts by filling missing data with NULL."""
    corrected_rows = []
    with open(file_path, 'r', encoding='utf-8') as f:
        header = f.readline()
        corrected_rows.append(header)
        for line in f:
            fields = line.strip().split(',')
            if len(fields) != expected_columns:
                if len(fields) < expected_columns:
                    fields += ['NULL'] * (expected_columns - len(fields))
                else:
                    fields = fields[:expected_columns]  
            corrected_rows.append(','.join(fields) + '\n')
    
    corrected_file = "corrected_hotels_data.csv"
    with open(corrected_file, 'w', encoding='utf-8') as f:
        f.writelines(corrected_rows)

    return corrected_file

# Detect expected number of columns
with open(file, 'r', encoding='utf-8') as f:
    header = f.readline()
    expected_columns = header.count(',') + 1

# Preprocess the file to handle inconsistent rows
corrected_file = preprocess_csv(file, expected_columns)

total_inserted = 0
for chunk in pd.read_csv(corrected_file, chunksize=chunk_size):
    inserted_count = 0
    with engine.connect() as conn:
        for _, row in chunk.iterrows():
            data = row.to_dict()

            # Filter data to match table columns
            filtered_data = {k: v for k, v in data.items() if k in grnc_table.columns}

            # Handle NaN values
            filtered_data = handle_nan_values(filtered_data)

            # Insert data into the database
            with conn.begin() as transaction:
                insert_statement = grnc_table.insert().values(filtered_data)
                conn.execute(insert_statement)
                inserted_count += 1

    total_inserted += inserted_count
    print(f"Inserted {inserted_count} records into the local database. Total inserted so far: {total_inserted}")

print("CSV data upload completed successfully.")
```
