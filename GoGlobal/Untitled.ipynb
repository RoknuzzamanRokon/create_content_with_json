```python
import requests
import xmltodict
import json

import requests

url = "https://gtr.xml.goglobal.travel/xmlwebservice.asmx"

payload = """<?xml version="1.0" encoding="utf-8"?>
    <soap12:Envelope xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance" 
                 xmlns:xsd="http://www.w3.org/2001/XMLSchema" 
                 xmlns:soap12="http://www.w3.org/2003/05/soap-envelope">
        <soap12:Body>
            <MakeRequest xmlns="http://www.goglobal.travel/">
                    <requestType>6</requestType>
                    <xmlRequest>
                        <![CDATA[
                            <Root>
                                <Header>
                                    <Agency>149548</Agency>
                                    <User>NOFSHONXMLTEST</User>
                                    <Password>W99IL98KY1G</Password>
                                    <Operation>HOTEL_INFO_REQUEST</Operation>
                                    <OperationType>Request</OperationType>
                                </Header>
                                <Main Version="2.2">
                                    <InfoHotelId>1194629</InfoHotelId>
                                    <InfoLanguage>en</InfoLanguage>
                                </Main>
                            </Root>
                        ]]>
                    </xmlRequest>
            </MakeRequest>
        </soap12:Body>
    </soap12:Envelope>"""


headers = {
          'Content-Type': 'application/soap+xml; charset=utf-8',
          'API-Operation': 'HOTEL_INFO_REQUEST'
        }



try:
    response = requests.post(url, data=payload, headers=headers)
    response.raise_for_status() 

    # Parse the XML response into a dictionary
    parsed_response = xmltodict.parse(response.text)
    
    # Convert the dictionary to JSON format
    json_response = json.dumps(parsed_response, indent=4)
    
    print("Response in JSON format:")
    print(json_response)
except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")
except Exception as e:
    print(f"An error occurred during parsing: {e}")
```

```python
import xml.etree.ElementTree as ET

# Extracting the MakeRequestResult field
xml_data = parsed_response["soap:Envelope"]["soap:Body"]["MakeRequestResponse"]["MakeRequestResult"]

# print(xml_data)

# Parsing the XML data
root = ET.fromstring(xml_data)

hotel_name = root.find(".//HotelName").text  
hotel_search_code = root.find(".//HotelSearchCode").text
hotel_id = root.find(".//HotelId").text
hotel_address = root.find(".//Address").text
hotel_citycode = root.find(".//CityCode").text
hotel_phone = root.find(".//Phone").text
hotel_fax = root.find(".//Fax").text
hotel_description = root.find(".//Description").text
print(hotel_description)
hotel_hotelFacilities = root.find(".//HotelFacilities").text
facilities_list = hotel_hotelFacilities.split("<BR />")

# Process each facility into the desired JSON structure
facilities_json = []
for facility in facilities_list:
    facility = facility.strip()  
    if facility:  
        facilities_json.append({
            "type": facility,
            "title": facility,
            "icon": "NULL"
        })


# print(f"hotel name = {facilities_json}")

# Proceess each room facility into content format.
hotel_roomFacilities = root.find(".//RoomFacilities").text
roomFacilities_list = hotel_roomFacilities.split("<BR />")

rm_facilities_json = []
for facility in roomFacilities_list:
    facility = facility.strip()
    if facility:
        rm_facilities_json.append({
            "type": facility,
            "title": facility,
            "icon": "NULL"
        })


# print(rm_facilities_json)


# This is picture section.
# hotel_picture = root.find(".//Pictures")
# print(hotel_picture)

pictures = root.find(".//Pictures")
# print(pictures)
# Initialize the photo_data list
photo_data = []


if pictures is not None:
    for picture in pictures.findall("Picture"):
        url = picture.text.strip() if picture.text else ""
        
        # Extract the Description attribute as title
        title = picture.attrib.get("Description", "")
        photo_data.append({
            "picture_id": "NULL",
            "title": title,
            "url": url
        })

# print(photo_data)

primary_photo_url = None  # Default to None

for photo in photo_data:
    if photo['title'] == 'Primary image':
        primary_photo_url = photo['url']
        break  

# print("Primary Photo URL:", primary_photo_url)
```
